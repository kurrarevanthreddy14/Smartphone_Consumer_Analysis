```python
df = pd.read_csv('8055_SurveyResponses.csv')
```

```python
df.head()
```

```python
import pandas as pd
from scipy.stats import ttest_ind

likert_map = {
    'Strongly Disagree': 1,
    'Disagree': 2,
    'Neutral': 3,
    'Agree': 4,
    'Strongly Agree': 5
}

likert_columns = [
    'Camera Quality', 'Battery Life', 'RAM Capacity', 'Display Size & Resolution',
    'Processor Performance', 'Internal Storage', 'OS & Interface', '5G Support',
    'Build Quality & Design', 'Software Updates'
]

for col in likert_columns:
    df[col] = df[col].map(likert_map)
```

```python
df['Brand_Switched'] = df['Have you changed your smartphone brand in your last purchase?'].str.strip()

ttest_results = {}

for feature in likert_columns:
    group_yes = df[df['Brand_Switched'] == 'Yes'][feature].dropna()
    group_no = df[df['Brand_Switched'] == 'No'][feature].dropna()

    # Perform t-test
    t_stat, p_val = ttest_ind(group_yes, group_no, equal_var=False)
    ttest_results[feature] = {'t_stat': round(t_stat, 3), 'p_value': round(p_val, 3)}

print("T-Test Results Comparing Brand Switchers and Non-Switchers:")
print("-" * 60)
for feature, result in ttest_results.items():
    print(f"{feature:<30} | t-stat: {result['t_stat']:<6} | p-value: {result['p_value']}")
```
